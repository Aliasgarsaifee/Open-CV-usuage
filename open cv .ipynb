```python
def to_rgb(temp):
    return (cv2.cvtColor(temp, cv2.COLOR_BGR2RGB))
```

```python
canvas = np.full((300, 300, 3), fill_value=255, dtype = "uint8")
```

```python
_=cv2.line(img=canvas, pt1=(0, 0), pt2=(300, 300), color=(0, 0, 255), thickness=5, lineType=cv2.LINE_AA)
```

```python
plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
```

```python
_=cv2.circle(img=canvas, center=(150, 150), radius=50, color=(255, 0, 0), thickness=-1, lineType=cv2.LINE_AA)
```

```python
_=cv2.rectangle(img=canvas, pt1=(0, 200), pt2=(100, 300), color=(0, 255, 0), thickness=5, lineType=cv2.LINE_AA)
```

```python
plt.imshow(to_rgb(canvas))
```

```python
image_resized = cv2.resize(image, (500,400))
plt.imshow(cv2.cvtColor(image_resized, cv2.COLOR_BGR2RGB))
```

```python
image_avg_blur=cv2.blur(image,(53,53))
plt.imshow(to_rgb(image_avg_blur))
```

```python
image_gauss_blur = cv2.GaussianBlur(image, (53, 53), 5)
```

```python
plt.imshow(cv2.cvtColor(image_gauss_blur, cv2.COLOR_BGR2RGB))
```

```python
image_gray_smooth = cv2.GaussianBlur(image_gray, (5, 5), 0)
canny = cv2.Canny(image_gray_smooth, threshold1=50, threshold2=80)
plt.imshow(canny, cmap='Greys')
```

```python
kernel = np.ones((3,3),np.uint8)
erosion = cv2.erode(canny,kernel,iterations = 1)
plt.imshow(to_rgb(erosion))
```

```python
dilation = cv2.dilate(canny,kernel,iterations = 1)
plt.imshow(to_rgb(dilation))
```

```python
#rotation of image by given degree
c=input(float)
M1=cv2.getRotationMatrix2D((rows/4,coloums/4),20,1)
falls_rot=cv2.warpAffine(falls,M1,(rows,coloums))
```
